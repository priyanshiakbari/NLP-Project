# Opening json file

```python
import json

# Loading json file
f = open('google_news.json')
data = json.load(f)

# Obtaining title and description dictonaries from json file
title_dict = data['title']
desc_dict = data['desc']

title_text = []
desc_text = []
# Retriving the text of title and description dictonaries and storing it into variables in string format
for key,value in title_dict.items():
    title_text.append(value)
    
for key,value in desc_dict.items():
    desc_text.append(value)
    
title_text = str(title_text)
desc_text = str(desc_text)

text = title_text + desc_text
text
```

# Lowercase

```python
text = text.lower()
text
```

# Removing Punctuations

```python
import string
print(string.punctuation)
```

```python
text = "".join([char for char in text if char not in string.punctuation+"’"+"‘"])
text
```

# Tokenization

```python
from nltk import word_tokenize

words = word_tokenize(text)
words
```

# Stopword Filtering

```python
from nltk.corpus import stopwords

stop_words = stopwords.words('english')

filtered_words = [word for word in words if word not in stop_words]
filtered_words
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer

porter = PorterStemmer()
stemmed = [porter.stem(word) for word in filtered_words]
```

# Part of Speech (POS) Tagger

```python
from nltk import pos_tag

pos = pos_tag(filtered_words)
pos
```

```python
len(filtered_words)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
  
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
lemmatized_words
```
